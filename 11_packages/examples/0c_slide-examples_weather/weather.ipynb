```python
rainy_days = weather[weather['Precipitation'] > 0]
# 1. for the weather dataframe,
# 2. in the preciptation column,
# 3. filter for values that are greater than zero

rainy_days
```

```python
# whereas min() and max() would return values, 
# idxmin() and idxmax() return instances/rows - "index min", "index max"

# whereas {df}.iloc[n] is index-based, similar to {list}[n] - "index location"
# {df}.loc is label-based - "location"


coldest = weather.loc[weather['Temperature'].idxmin()]
coldest
```

```python
# let's plot that dataset out onto a graph

import matplotlib.pyplot as plt

# plt.plot(x_axis, y_axis, **kwargs)
plt.plot(weather['Date'], weather['Temperature'], marker='o')

# configuring the graph
# see docs: https://matplotlib.org/stable/api/pyplot_summary.html
plt.title('Daily Precipitation')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.xticks(rotation=45)
plt.tight_layout()

# ok let's see
plt.show()
```
